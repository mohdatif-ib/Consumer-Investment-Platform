# 03 Startup Scoring

Rank startups using a VC-style weighted scoring model.

```python
from pathlib import Path
import sys
import pandas as pd

PROJECT_ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
sys.path.append(str(PROJECT_ROOT / 'src'))

from consumer_brand_intel.startup_scoring import build_scoring_features, score_startups

processed_dir = PROJECT_ROOT / 'data' / 'processed'
events = pd.read_csv(processed_dir / 'funding_events.csv')
investments = pd.read_csv(processed_dir / 'investments.csv')
investors = pd.read_csv(processed_dir / 'investors.csv')
```

```python
features = build_scoring_features(events, investments, investors)
features.head()
```

```python
rankings = score_startups(features)
rankings[['rank', 'company_name', 'sector', 'total_funding', 'investment_attractiveness_score']].head(50)
```

## Model Assumptions

- Funding growth is used as a proxy for investor demand and market traction.
- Investor quality is inferred from observed investment activity and portfolio breadth.
- Sector growth captures category momentum.
- Funding consistency rewards repeatable capital formation patterns.
