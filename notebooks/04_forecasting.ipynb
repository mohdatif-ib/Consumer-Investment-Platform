# 04 Forecasting

Forecast the next 12 months of funding activity with Linear Regression and ARIMA.

```python
from pathlib import Path
import sys
import pandas as pd
import plotly.express as px

PROJECT_ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
sys.path.append(str(PROJECT_ROOT / 'src'))

from consumer_brand_intel.forecasting import load_monthly_funding, forecast_with_linear_regression, forecast_with_arima, identify_fastest_growing_sectors

processed_dir = PROJECT_ROOT / 'data' / 'processed'
funding_events_path = processed_dir / 'funding_events.csv'
```

```python
monthly = load_monthly_funding(funding_events_path)
monthly.head()
```

```python
linear_forecast = forecast_with_linear_regression(monthly, periods=12)
arima_forecast = forecast_with_arima(monthly, periods=12)
forecasts = pd.concat([linear_forecast, arima_forecast], ignore_index=True)
px.line(forecasts, x='forecast_month', y='predicted_funding', color='model', markers=True, title='Next 12 Months Funding Forecast')
```
