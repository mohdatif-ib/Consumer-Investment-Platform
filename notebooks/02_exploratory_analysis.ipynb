# 02 Exploratory Analysis

Analyze consumer brand funding by sector, city, investor, funding stage, and month.

```python
from pathlib import Path
import pandas as pd
import plotly.express as px

PROJECT_ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
processed_dir = PROJECT_ROOT / 'data' / 'processed'

events = pd.read_csv(processed_dir / 'funding_events.csv', parse_dates=['funding_date'])
investments = pd.read_csv(processed_dir / 'investments.csv')
investors = pd.read_csv(processed_dir / 'investors.csv')
```

```python
sector_funding = events.groupby('sector', as_index=False)['amount_usd'].sum().sort_values('amount_usd', ascending=False)
px.bar(sector_funding, x='sector', y='amount_usd', title='Funding by Sector')
```

```python
city_funding = events.groupby('city', as_index=False)['amount_usd'].sum().sort_values('amount_usd', ascending=False)
px.bar(city_funding, x='city', y='amount_usd', title='Funding by City')
```

```python
monthly = events.set_index('funding_date').resample('MS')['amount_usd'].sum().reset_index()
px.line(monthly, x='funding_date', y='amount_usd', markers=True, title='Funding Over Time')
```

```python
investor_activity = investments.merge(investors, on='investor_id').groupby('investor_name', as_index=False).agg(rounds=('round_id', 'nunique'), total_invested=('amount', 'sum')).sort_values('rounds', ascending=False)
px.bar(investor_activity, x='investor_name', y='rounds', title='Investor Activity')
```

```python
stage_distribution = events.groupby('funding_stage', as_index=False)['event_id'].nunique().rename(columns={'event_id': 'rounds'})
px.pie(stage_distribution, names='funding_stage', values='rounds', title='Funding Stage Distribution')
```
